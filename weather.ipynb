```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('weatherAUS.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data["Date"] = pd.to_datetime(data["Date"], format = "%Y-%m-%d", errors = "coerce")
```

```python
data
```

```python
data["RainToday"] = pd.get_dummies(data["RainToday"], drop_first = True)
data["RainTomorrow"] = pd.get_dummies(data["RainTomorrow"], drop_first = True)
data
```

```python
data.dtypes
```

```python

data["Date"]=pd.to_numeric(data["Date"], downcast="float")
```

```python
data
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
data['Location'] = le.fit_transform(data['Location'])
data['WindDir9am'] = le.fit_transform(data['WindDir9am'])
data['WindDir3pm'] = le.fit_transform(data['WindDir3pm'])
data['WindGustDir'] = le.fit_transform(data['WindGustDir'])
```

```python
X= data.iloc[:, :22]
print(X)
```

```python
Y= data.iloc[:, -1]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
LinearRegressionScore = lr.score(X_test,Y_test)
print("Accuracy obtained by Linear Regression model:",LinearRegressionScore*100)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(booster = 'gbtree', learning_rate = 0.1, max_depth = 5, n_estimators = 180)
xgb.fit(X_train, Y_train);
```

```python
XGBClassifierScore = xgb.score(X_test,Y_test)
print("Accuracy obtained by XGB Classifier model:", XGBClassifierScore*100)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(X_train, Y_train)
```

```python
GradientBoostingClassifierScore = gb.score(X_test,Y_test)
print("Accuracy obtained by Gradient Boosting Classifier model:",GradientBoostingClassifierScore*100)
```
